```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def fun(x):
  y = (x*x) + 5
  return y
```

```python
t1 = np.arange(-1, 1, 0.1)

plt.plot(fun(t1))
plt.xlabel('x')
plt.ylabel('y')
plt.title('wartość funkcji dla x z przedziału (-1, 1)')
```

```python
t2 = numpy.arange(-6, 6, 0.1)

plt.plot(fun(t2))
plt.xlabel('x')
plt.ylabel('y')
plt.title('wartość funkcji dla x z przedziału (-6, 6)')
```

```python
t3 = numpy.arange(0, 5, 0.1)

plt.plot(fun(t3))
plt.xlabel('x')
plt.ylabel('y')
plt.title('wartość funkcji dla x z przedziału (0, 5)')
```
